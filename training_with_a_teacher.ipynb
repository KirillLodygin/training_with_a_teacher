## Описание проекта

Интернет-магазин «В один клик» разрабатывает программу действий по удержанию активности постоянных клиентов. Сделать это можно с помощью персонализированных предложений. Необходимо найти решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

Для этого нужно:
1. Промаркировать уровень финансовой активности постоянных покупателей. В компании принято выделять два уровня активности: «снизилась», если клиент стал покупать меньше товаров, и «прежний уровень».

2. Собрать данные по клиентам по следующим группам:
   - признаки, которые описывают коммуникацию сотрудников компании с клиентом;
   - признаки, которые описывают продуктовое поведение покупателя, какие товары покупает и как часто;
   - признаки, которые описывают покупательское поведение клиента, сколько тратил в магазине;
   - признаки, которые которые описывают поведение покупателя на сайте, как много страниц просматривает и сколько времени проводит на сайте.
   
3. Построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца. В исследование включить дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца. Исследуем показатели стобца "Покупательская активность" датафрейма market_file.csv.
Создадим функцию для отрисовки круговой диограммы и применим ее к столбцу "Покупательская активность".

````python
def drew_pie_chart(df, column, title):
    current_df = df.groupby(column)
    vals = (current_df.size() / len(df)).tolist()
    labels = current_df[column].unique()
    
    service_list = [0.0]
    
    for i in range(len(vals) - 1):
        service_list.append(0.1)
    
    explode = tuple(service_list)

    fig, ax = plt.subplots()
    ax.pie(vals, 
       labels=labels, 
       autopct='%1.1f%%', 
       shadow=True, 
       explode=explode, 
       wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"}, 
       rotatelabels=False)
    ax.axis("equal")
    plt.title(title)
    plt.show()
    
drew_pie_chart(market_file, 
               "Покупательская активность", 
               "Соотношение групп по показателю 'Покупательская активность'")    
````

Покупательская активность снизилась почти у 40% покупателей.
Посмотрим, какие категории товаров предпочитают покупать люди со снизившейся покупательской активностью.

````python
buyers_with_decreased_activity = market_file.query("`Покупательская активность` == 'Снизилась'")
drew_pie_chart(buyers_with_decreased_activity, 
               'Популярная_категория', 
               'Популярные товары у покупателей со снижающейся активностью')
````

````python
def get_values_by_groups(df, condition, group_field, result_field):
    result_df = (df.query(condition))
    result = result_df.groupby(group_field)[result_field].median()
    print(result)
````

##  Объединение таблиц и подготовка к корреляционному анализу

````python
def get_new_df_columns(df, category, value):
    pivoted_df = df.pivot_table(index='id', columns=category, values=value).reset_index()
    
    return pivoted_df

def rename_columns(df, prefix):
    columns_to_rename = {col: f'{prefix}{col}' for col in df.columns if col != 'id'}
    df.rename(columns=columns_to_rename, inplace=True)
    return df
````

##  Корреляционный анализ

Построим матрицу корреляций, предварительно нормировав количественные признаки и кодировав категориальные.

````python
def min_max_normalize(df):
    result = df.copy()
    numeric_df = df.select_dtypes(include=['number'])
    columns = numeric_df.columns.tolist()
    for column in columns:
        max_value = numeric_df[column].max()
        min_value = numeric_df[column].min()
        numeric_df[column] = (numeric_df[column] - min_value) / (max_value - min_value)
    
    result[columns] = numeric_df[columns]
    return result
````

###  Выделим признаки для кодирования и масштабирования

````python
X_train, X_test, y_train, y_test = train_test_split(
    final_merged_df.drop(['Покупательская активность'], 
        axis=1
    ),
    final_merged_df['Покупательская активность'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = final_merged_df['Покупательская активность']
)

ohe_columns = ['Популярная_категория', 'Тип сервиса', 'Разрешить сообщать']
ord_columns = ['Средний_просмотр_категорий_за_визит', 'Маркет_актив_тек_мес']

num_columns =[
	'market_time_предыдущий_месяц',
    'market_time_текущий_месяц',
	'Страниц_за_визит',
    'Неоплаченные_продукты_штук_квартал',
    'Ошибка_сервиса',
    'Маркет_актив_6_мес',
	'Акционные_покупки',
	'Длительность',
    'market_money_предыдущий_месяц',
    'market_money_препредыдущий_месяц',
    'market_money_текущий_месяц'
]
````

````python
ohe_pipe = Pipeline(
    [
        ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(handle_unknown='ignore', drop='first'))
    ]
)

ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['1', '2', '3', '4', '5', '6'],
                    ['3', '4', '5'],
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)

data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns),
    ], 
    remainder='passthrough'
)
````

<font color='blue'><b>Комментарий ревьюера: </b></font> ❌\
<font color='red'>~~Для линейных моделей обязательно указывать параметр `drop='first'` в OneHotEncoder. Например, есть у тебя некий бинарный признак. Без дропа у тебя создастся 2 колонки, причём всегда в одной будет 0, в другой - 1, потому что признак не может принимать сразу 2 значения. Математически можно записать, что признак1 + признак2 = 1, либо признак1 + признак2 - 1 = 0. Говоря терминами математики, это линейно зависимые признаки. А линейно зависимые признаки - это мультиколлнеарность, явление, мешающее обучение линейных моделей. Так что добавь, дроп, пожалуйста.~~</font>
